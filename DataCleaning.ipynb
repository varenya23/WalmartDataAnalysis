```python
#for duplicate values
df.duplicated().sum()
```

```python
#for missing values
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
#This will return error, beacuse of the $ sign palced at the start of the string
df['unit_price'].astype(float)
```

```python
df['unit_price'] = df['unit_price'].str.replace('$', '').astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df['total_amount'] = df['unit_price'] * df['quantity']
df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.to_csv('clean_walmat_data', index=False)
```
